## Outlier Detection And  Method Selection

```python
numeric_col=df.select_dtypes(include=['int64','float64']).columns
numeric_col=numeric_col.drop('Coupon_Accepted')
numeric_col
```

### skewness

```python
for col in numeric_col:
    df[col].skew()
    print('skewness',df[col])
```

### Skewness is approximately 0. It indicates symmetrical or normal distribution

### Graphical representation of outlier detection of numerical columns

```python
fig, axes=plt.subplots(nrows=4,ncols=2,figsize=(10,10))
axes=axes.flatten()
for i,column in enumerate(numeric_col):
    sns.boxplot(y=df[column], ax=axes[i])
    axes[i].set_title(column)
```

### IQR method

```python
for col in numeric_col:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q1-q3
    lower_level=q1-1.5*iqr
    upper_level=q3+1.5*iqr
    outliers = df[(df[col] < lower_level) | (df[col] > upper_level)]
outliers
```

### Z-Score Method

```python
for col in numeric_col:
    upper_limit=df[col].mean()+3*df[col].std()
    lower_limit=df[col].mean()-3*df[col].std()
    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
outliers
```

### Considered both IQR(inter quartile range) and Z-score method to identify outliers. IQR detected 2724 outliers while Z-score detect none. Z-score identifies no outliers means that most data points falls within a typical range for normal distribution. There is no extreme value inm this dataset according to z-sore method. Because the data distribution is symmetrical i decides to use Z-score method

```python
encoded_data = pd.get_dummies(data=df,columns=['destination','passanger','weather','time','coupon','expiration','gender','age','MartialStatus',
'education','occupation','income','Bar','CoffeeHouse','CarryAway','RestaurantLessThan20','Restaurant20To50'])
encoded_data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
reduced_data = pca.fit_transform(encoded_data)
reduced_data
new_df = pd.DataFrame(reduced_data)
new_df
```

```python
new_df['Coupon_Accepted']=df['Coupon_Accepted']
X = new_df.drop('Coupon_Accepted',axis=1)
y = new_df['Coupon_Accepted']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
rf = RandomForestClassifier()
model = rf.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
print(cm)
```

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression()
l_model = Lr.fit(X_train_scaled, y_train)
y_pred = l_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.svm import SVC
sv = SVC()
sv_model = sv.fit(X_train_scaled,y_train)
y_pred = sv_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
#Classification model (Random Forest) with hyperparameter tuning
from sklearn.model_selection import train_test_split, GridSearchCV
param_grid_classification = {'n_estimators': [100, 200, 300], 'max_depth': [None]}
grid_search_classification = GridSearchCV(RandomForestClassifier(), param_grid=param_grid_classification,cv=5)
grid_search_classification.fit(X_train_scaled, y_train)
best_params_classification = grid_search_classification.best_params_
best_model_classification = grid_search_classification.best_estimator_
y_pred_classification = best_model_classification.predict(X_test)
accuracy_classification = accuracy_score(y_test, y_pred_classification)
print("Classification Accuracy:", accuracy_classification)
```
